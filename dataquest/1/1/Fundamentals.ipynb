1. Intro
Helper functions to operate on data:

```python
from csv import reader

def load_data(file_name):
    file = open(file = file_name, encoding = 'utf8')
    data = reader(file)
    return list(data)


def explore_data(dataset, start, end, rows_and_columns=False):
    dataset_slice = dataset[start:end]    
    for row in dataset_slice:
        print(row)
        print('\n') # adds a new (empty) line after each row

    if rows_and_columns:
        print('Number of rows:', len(dataset))
        print('Number of columns:', len(dataset[0]))
```

2. Data Loading 
From files:

- Apple Apps https://www.kaggle.com/ramamet4/app-store-apple-data-set-10k-apps/home

```python
apple_data = load_data('AppleStore.csv')
google_data = load_data('googleplaystore.csv')
explore_data(apple_data, 0, 2, True)
explore_data(google_data, 0, 2, True)
```

3. Data cleaning
Need to remove:
 - invalid
 - duplicates
 - Non-English
 - Payed

```python
explore_data(google_data, 10473,10474)
# Deleting invalid row - only once!
del google_data[10473]
```

```python
google_headers = google_data[0]
apple_headers = apple_data[0]
# Removing headers
google_data = google_data[1:]
apple_data = apple_data[1:]

# duplicates
duplicate_apps = []
unique_apps = []

for app in google_data:
    name = app[0]
    if name in unique_apps:
        duplicate_apps.append(name)
    else:
        unique_apps.append(name)
    
print('Number of duplicate apps:', len(duplicate_apps))
print('\n')
print('Examples of duplicate apps:', duplicate_apps[:15])

reviews_max = {}

for app in google_data:
    name = app[0]
    n_reviews = float(app[3])
    
    if name in reviews_max and reviews_max[name] < n_reviews:
        reviews_max[name] = n_reviews
        
    elif name not in reviews_max:
        reviews_max[name] = n_reviews
        
google_data_clean = []
already_added = []

for app in google_data:
    name = app[0]
    n_reviews = float(app[3])
    
    if (reviews_max[name] == n_reviews) and (name not in already_added):
        google_data_clean.append(app)
        already_added.append(name) # make sure this is inside the if block
        
explore_data(google_data_clean, 0, 3, True)
```

```python
# Removing non-English
def is_english(string):
    non_ascii = 0
    for character in string:
        if ord(character) > 127:
            non_ascii += 1
    
    if non_ascii > 3:
        return False
    else:
        return True

google_english = []
apple_english = []

for app in google_data_clean:
    name = app[0]
    if is_english(name):
        google_english.append(app)
        
for app in apple_data:
    name = app[2]
    if is_english(name):
        apple_english.append(app)
        
explore_data(google_english, 0, 3, True)
print('\n')
explore_data(apple_english, 0, 3, True)
```

```python
# Remove paid
google_free = []
google_paid = []
apple_free = []
apple_paid = []

for app in google_english:
    if app[7] == '0':
        google_free.append(app)
    else:
        google_paid.append(app)

for app in apple_english:
    if float(app[5]) <= 0:
        apple_free.append(app)
    else:
        apple_paid.append(app)

explore_data(google_free, 0, 3, True)
print('\n')
explore_data(google_paid, 0, 3, True)
print('\n')
explore_data(apple_free, 0, 3, True)
print('\n')
explore_data(apple_paid, 0, 3, True)
```

4. Buidling Frequency tables

```python
# Genre Popularity
def freq_table(dataset, index):
    table = {}
    total = 0
    
    for row in dataset:
        total += 1
        value = row[index]
        if value in table:
            table[value] += 1
        else:
            table[value] = 1
    
    table_percentages = {}
    for key in table:
        percentage = (table[key] / total) * 100
        table_percentages[key] = percentage 
    
    return table_percentages

def display_table(dataset, index):
    table = freq_table(dataset, index)
    table_display = []
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)
        
    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ':', entry[0])
        
print("Google Genres:")
display_table(google_free, 1)
print("\nApple Genres:")
display_table(apple_free, -5)
```

```python
# Most Popular Apps by Genre on the App Store
genres_apple = freq_table(apple_free, -5)

for genre in genres_apple:
    total = 0
    len_genre = 0
    for app in apple_free:
        genre_app = app[-5]
        if genre_app == genre:            
            n_ratings = float(app[6])
            total += n_ratings
            len_genre += 1
    avg_n_ratings = total / len_genre
    print(genre, ':', avg_n_ratings)
```

```python
# Most Popular Apps by Genre on Google Play
categories_google = freq_table(google_free, 1)

for category in categories_google:
    total = 0
    len_category = 0
    for app in google_free:
        category_app = app[1]
        if category_app == category:            
            n_installs = app[5]
            n_installs = n_installs.replace(',', '')
            n_installs = n_installs.replace('+', '')
            total += float(n_installs)
            len_category += 1
    avg_n_installs = total / len_category
    print(category, ':', avg_n_installs)
```
